### Data Preparation

* Droping unnecessary columns using drop function

```python
dataset.drop(['id','your_favoritearticle_today','url','latitude','longitude', 'officeaddress','homeaddress','zip_code','pymnt_plan',
              'YearsCodingProf', 'orignal_website_directory','Process Date','Process Date','Posting Updated',
             'website','dateAdded','phones','Time'],axis=1,inplace=True)
dataset.head(2)
```

 * loan amount is in string formate converting that in to numeric 

```python
dataset['LOAN_AMT'] = dataset.LOAN_AMT.str[1:]
dataset['LOAN_AMT'] = (dataset['LOAN_AMT'].str.split(',', n=1, expand = True)[0] + dataset['LOAN_AMT'].str.split(',', n=1, expand = True)[1]).astype('float')
```

* posting date column is in string formate converting that in to date time format using 'to_datetime'

```python
import datetime as dt
```

```python
dataset['Posting Date'] = pd.to_datetime(dataset['Posting Date'])
dataset['Posting_Date'] =dataset['Posting Date'].dt.date;
dataset['Posting_Date_year'] = (dataset['Posting Date'].dt.month)
dataset['Posting_Date_year']
```

* companysize column is in string value we extract the size of the company and converting into numeric value using 'to_numeric'

```python
dataset['CompanySize'] =dataset['CompanySize'].str.split(" ", expand =True)[2]
dataset['CompanySize'] =dataset['CompanySize'].str.split(',', expand = True)[0]+dataset['CompanySize'].str.split(',', expand = True)[1].replace(np.nan, str(' '))
dataset['CompanySize'] = dataset['CompanySize'].replace('more ', 1000)
dataset['CompanySize'] = pd.to_numeric(dataset['CompanySize'])
```

* from string value extract the year of coding and converting that in to numeric

```python
dataset['YearsCoding']
dataset['YearsCoding'] = dataset['YearsCoding'].str.split('-| | ', expand = True)[0]
dataset['YearsCoding'] = pd.to_numeric(dataset['YearsCoding'])
```

```python
dataset['last_pymnt_month'] = dataset['last_pymnt_d'].str.split('-', expand = True)[1]
dataset['last_pymnt_month']
```

```python
dataset['earliest_cr_line_month'] = dataset['earliest_cr_line'].str.split('-', expand = True)[1]
dataset['last_credit_pull_month'] = dataset['last_credit_pull_d'].str.split('-', expand = True)[1]
dataset['last_credit_pull_month']
```

* convrting term to numeric and time to datatime format

```python
dataset['term'] = dataset['term'].str.split(' ', expand = True)[1]
dataset['term'] = pd.to_numeric(dataset['term'])
```

```python
dataset.head(3)
```

* filling the nan(not a number) values 
* for categorical data using 'mode'
* for numerical data using 'median'

```python
dataset['Minimum Qual Requirements'].fillna('missing', inplace =True)
dataset['Preferred Skills'].fillna('missing', inplace =True)
dataset['To Apply'].fillna('missing', inplace =True)
dataset['Residency Requirement'].fillna('missing', inplace =True)
dataset['Posting Date'].fillna(pd.to_datetime('2019-11-01'), inplace =True)
dataset['UndergradMajor'].fillna('Computer science, computer engineering, or software engineering', inplace =True)

dataset['YearsCoding'].fillna(dataset['YearsCoding'].median(), inplace =True)
dataset['DevType'].fillna('Full-stack developer', inplace =True)
dataset['HopeFiveYears'].fillna("Working in a different or more specialized technical role than the one I'm in now", inplace =True)
dataset['JobSearchStatus'].fillna('I’m not actively looking, but I am open to new opportunities', inplace =True)
dataset['LastNewJob'].fillna('Less than a year ago', inplace =True)
dataset['LanguageWorkedWith'].fillna('C#;JavaScript;SQL;HTML;CSS', inplace =True)
dataset['IDE'].fillna('Visual Studio;Visual Studio Code', inplace =True)
dataset['OperatingSystem'].fillna('Windows', inplace =True)
dataset['title'].fillna('Debt consolidation', inplace =True)
dataset['FormalEducation'].fillna('Bachelor’s degree (BA, BS, B.Eng., etc.)', inplace =True)
dataset['Posting_Date_year'].fillna(11, inplace =True)
```

### Data Modeling

```python
dataset.drop(['Posting Date','earliest_cr_line','last_pymnt_d','last_credit_pull_d','CrimeTime', 'Posting_Date'],axis=1,inplace=True)
dataset.head(2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix, accuracy_score, r2_score
from sklearn.ensemble import RandomForestClassifier 
from sklearn.ensemble import RandomForestRegressor
```

```python
X = dataset.iloc[:, 2:]
y1 = dataset['Target_Salary']
y2 = dataset['Target_Satisfied']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y1, test_size=0.20, random_state=42)
X2_train, X2_test, y2_train, y2_test = train_test_split(X, y2, test_size=0.20, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

##### using RandomForestClassifier we are going to classifing the Target_Satisfied

```python
model = RandomForestClassifier(n_estimators=200, max_depth= 4, min_samples_split=7, min_samples_leaf = 1, random_state = 1)
```

### SMOTE 

* using somte technique we are going to balance the imbalance the data

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
counter = Counter(y2_train)
print('before smote: ',counter)
smote = SMOTE(k_neighbors=5)
X_smote, y_smote = smote.fit_resample(X2_train, y2_train.ravel())
counter = Counter(y_smote)
print('after smote: ',counter)
```

```python
model = model.fit(X_smote, y_smote)
```

```python
predict_train = model.predict(X2_train)
predict_test = model.predict(X2_test)
print(classification_report(y2_test, predict_test))
```

```python
confusion_matrix(y2_test, predict_test)
```

```python
accuracy_score(y2_test, predict_test)
```
